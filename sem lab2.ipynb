```python

import xml.etree.cElementTree as ET
import lxml.etree
from xml.etree.ElementTree import Element, SubElement, ElementTree
from xml.dom.minidom import parseString


def edit(root):
    for name in root.iter("company_name"):
        if name.text == 'Стратоплан, онлайн-школа менеджеров':
            name.text = 'Стратоплан'
    for city in root.iter('city'):
        if city.text == 'Бровари':
            city.text = 'Київська обл.'
    for salary in root.iter('salary'):
        if salary.text=='10000–13000':
            salary.text='10000'
        if salary.text=='50000–70000':
            salary.text='50000'
        if salary.text=='15000–20000':
            salary.text='15000'
        if salary.text=='8000–12000':
            salary.text='8000'
    for city in root.iter('city'):
        city.attrib['country']='Ukraine'
    for salary in root.iter('salary'):
        del salary.attrib['currency']
    work = SubElement(root, 'work')
    work.set('id', 'element ' + str(14))
    job = SubElement(work, 'job')
    job.text = 'Python developer'
    company_name = SubElement(work, 'company_name')
    company_name.text = 'ZeTeam'
    salary = SubElement(work, 'salary')
    salary.text = str(18000)
    city = SubElement(work, 'city')
    city.text = 'Херсон'
    city.set('country','Ukraine')
    return root
        
def prettify(tree, filename):
    with open(filename, "wb") as f:
        tree.write(f)
    with open(filename, 'r', encoding='utf-8') as file:
        xml_text = file.read()
        prettified_xml = parseString(xml_text).toprettyxml()
        file.close()
    with open(filename, 'w', encoding='utf-8') as file:
        
        file.write(prettified_xml)
        file.close()

def newhtml():
    dom = lxml.etree.parse('workuaaaa.xml')
    xslt = lxml.etree.parse('workua.xslt')
    transform = lxml.etree.XSLT(xslt)
    newdom = transform(dom)
    with open('work.html', 'w') as file:
        file.write(str(newdom))
        file.close()


tree = ET.ElementTree(file="workuaaa.xml")
root = tree.getroot()
root = edit(root)
tree = ET.ElementTree(root)
prettify(tree, 'workuaaaa.xml')
newhtml()       
```
